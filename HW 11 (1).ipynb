(a) Classification Decision Tree Overview and Applications
What it does:
A Classification Decision Tree solves problems where the goal is to classify data into specific categories. It splits data step by step into smaller groups, based on specific conditions about the features (like "Is age greater than 30?"). The process continues until a final category (or class) is reached at the end of the tree.

Real-world examples:

Email spam detection: Classifying emails as "spam" or "not spam."
Medical diagnosis: Predicting whether a patient has a disease based on symptoms or test results.
Fraud detection: Classifying transactions as "fraudulent" or "legitimate."
Customer segmentation: Grouping customers into categories like "loyal" or "at risk."
(b) Difference Between Classification Decision Trees and Multiple Linear Regression
Classification Decision Tree:

A decision tree predicts a category (e.g., spam or not spam) by making sequential decisions based on features.
At each step, the data is split based on a condition (e.g., "Is age greater than 30?"), and this continues until the data reaches a final class (e.g., "spam").
The output is a category (e.g., "Yes" or "No").
Multiple Linear Regression:

Multiple linear regression predicts a continuous value by fitting a straight-line relationship to the data.
It uses the equation:
y = b0 + b1 * x1 + b2 * x2 + ... + bn * xn
where:
y = predicted value,
b0 = intercept (a constant),
b1, b2, ..., bn = weights (coefficients) for each input feature,
x1, x2, ..., xn = input feature values.
The output is a numerical value (e.g., predicting a house price or salary).

2,
Accuracy
Formula: Accuracy = (TP + TN) / (TP + TN + FP + FN)
What it measures: Accuracy tells us the proportion of correct predictions (both true positives and true negatives) out of the total predictions made by the model.
When to use: Accuracy is most useful when the classes are balanced, meaning there is a similar number of positive and negative cases.
Example: Measuring the overall performance of an email spam classifier when there are equal numbers of spam and non-spam emails.

Sensitivity
Formula: Sensitivity = TP / (TP + FN)
What it measures: Sensitivity, also called recall or the true positive rate, measures how well the model identifies actual positives.
When to use: Sensitivity is important when it is critical to catch all positive cases, even if it means increasing false positives.
Example: Diagnosing diseases where missing a positive case (false negative) could have serious consequences.

Specificity
Formula: Specificity = TN / (TN + FP)
What it measures: Specificity measures how well the model identifies actual negatives. It focuses on avoiding false positives.
When to use: Specificity is important when false positives must be minimized, such as when additional follow-up tests are expensive or invasive.
Example: A cancer screening test where false positives could lead to unnecessary and costly procedures.

Precision
Formula: Precision = TP / (TP + FP)
What it measures: Precision measures how many of the predicted positives are actually correct. It tells us the reliability of a positive prediction.
When to use: Precision is critical when the cost of false positives is high, such as when a wrong prediction could cause inconvenience or harm.
Example: Fraud detection systems, where incorrectly labeling legitimate transactions as fraudulent could frustrate users.

```python
8,
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

# Reload and preprocess the dataset
file_path = "/mnt/data/amazonbooks.csv"
ab = pd.read_csv(file_path, encoding="ISO-8859-1")

# Remove specified columns
ab_cleaned = ab.drop(columns=['Weight_oz', 'Width', 'Height'])

# Drop rows with missing values
ab_cleaned = ab_cleaned.dropna()

# Convert 'Pub year' and 'NumPages' to integers
ab_cleaned['Pub year'] = ab_cleaned['Pub year'].astype(int)
ab_cleaned['NumPages'] = ab_cleaned['NumPages'].astype(int)

# Convert 'Hard_or_Paper' to a categorical type
ab_cleaned['Hard_or_Paper'] = ab_cleaned['Hard_or_Paper'].astype('category')

# Prepare X and y for clf2
X_clf2 = ab_cleaned[['NumPages', 'Thick', 'List Price']]
y = pd.get_dummies(ab_cleaned['Hard_or_Paper'], drop_first=True)  # Encode 'H' as 1 and others as 0

# Train/test split
X_train_clf2, X_test_clf2, y_train, y_test = train_test_split(X_clf2, y, test_size=0.2, random_state=42)

# Train clf2 (max_depth=4)
clf2 = DecisionTreeClassifier(max_depth=4, random_state=42)
clf2.fit(X_train_clf2, y_train)

# Step 1: Get feature importances
feature_importances = clf2.feature_importances_

# Step 2: Get feature names
feature_names = clf2.feature_names_in_

# Step 3: Combine feature names with their importance
importances_df = pd.DataFrame({
    "Feature": feature_names,
    "Importance": feature_importances
}).sort_values(by="Importance", ascending=False)

# Display feature importances table
print(importances_df)

# Visualize feature importances
plt.figure(figsize=(8, 5))
plt.bar(importances_df["Feature"], importances_df["Importance"], color="skyblue")
plt.xlabel("Feature")
plt.ylabel("Importance")
plt.title("Feature Importance for clf2")
plt.show()

```

In a linear regression model, coefficients represent the direct impact of a one-unit change in a predictor variable on the dependent variable, assuming all other variables remain constant. The interpretation is straightforward and linear, reflecting additive contributions to the outcome.

In contrast, for decision trees, feature importances measure how much each predictor variable reduces uncertainty (e.g., Gini impurity or entropy) across all splits in the tree. Rather than direct numerical impacts, these importances show relative contributions to the tree's predictive power, with no explicit additive or linear relationship. Decision trees provide a holistic view of feature influence across multiple splits, which may not directly map to real-world units.
