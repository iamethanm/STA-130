```python
3
print("Simple Summaries of the Columns:")
print(df.info())
print("Summary Statistics for Numeric Columns:")
print(df.describe())
for column in df.columns:
    print(f"Value Counts for Column '{column}':")
    print(df[column].value_counts())
    print("\n")
```

```python
4
The df.shape attribute returns a tuple representing the dimensions of the DataFrame. It shows the number of rows and columns in the dataset.
The df.describe() method provides summary statistics for numeric columns by default. It includes metrics like count, mean, standard deviation, min, and max values.
Non-Numeric Variables: df.describe() only analyzes numeric columns by default. If your dataset has non-numeric variables (e.g., strings, dates), these columns will not be included in the summary statistics. This can lead to fewer columns being reported in df.describe() compared to the total number of columns indicated by df.shape.
Missing Values: The “count” value in df.describe() represents the number of non-null entries for each numeric column. If there are missing values (NaNs) in any numeric column, the count will be less than the total number of rows reported by df.shape. This discrepancy occurs because df.describe() only counts non-null values.
```

```python
5
Attribute: A property of a DataFrame that provides information without performing any operations, accessed without parentheses. Example: df.shape.
Method: A function associated with a DataFrame that performs operations and often requires parentheses. Example: df.describe().
        
```

```python
6
Count: The number of non-null (non-missing) values in the column. This helps you understand how many valid data points are available for analysis.
Mean: The average value of the column. It is calculated by summing all the non-null values and dividing by the count of those values.
Std (Standard Deviation): A measure of the amount of variation or dispersion in the values. A low standard deviation indicates that the values tend to be close to the mean, while a high standard deviation indicates that the values are spread out over a wider range.
Min: The minimum value in the column. This is the smallest value among the non-null entries.
25% (First Quartile): The value below which 25% of the data falls. It is also known as the first quartile or Q1.
50% (Median): The middle value of the column when the values are sorted in ascending order. It is also known as the second quartile or Q2. Half of the data points are below this value, and half are above.
75% (Third Quartile): The value below which 75% of the data falls. It is also known as the third quartile or Q3.
Max: The maximum value in the column. This is the largest value among the non-null entries.
```

```python
7
print("Before Removing Missing Data:")
print(df.info())
del df['column_name']
df = df.dropna()
print("After Removing Missing Data:")
print(df.info())
>>>Importance of Applying del df['col'] Before df.dropna():
When both methods are used together, applying del df['col'] before df.dropna() can be important because it ensures that columns with a high percentage of missing values are removed first. This can prevent the unnecessary removal of rows that might have been complete if the problematic column was deleted first.
Justification:
Before: The dataset might have several columns with missing values, and some rows might be incomplete.
After: By removing columns with a high percentage of missing values first, we ensure that we retain as many complete rows as possible. Then, by using df.dropna(), we remove any remaining rows with missing values, resulting in a clean dataset.
>>>
```
